```python
# split with noisy data
x , y = make_moons(n_samples = 300 , shuffle = True , noise = 0.5)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.3)
```

```python
x_train.shape
```

```python
model = MLPClassifier(hidden_layer_sizes = (5,) , activation = 'logistic' , solver = 'adam' , max_iter = 10000 , verbose = 10 , tol = 0.00001)
model.fit(x_train , y_train)
```

```python
predict = model.predict(x_test)
print(predict)
```

```python
print(model.score(x_test , y_test))
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test , predict)
ac
```

```python
Test= plt.scatter(x_test[:,0] , y_test)
Predict = plt.scatter(x_test[:,0],predict)

plt.legend([Predict,Test],["Predicted Data" , "Real Data"])
plt.show()
```
